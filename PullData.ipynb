## Pull Data With SODA API Client

```python
import os
from sodapy import Socrata

APP_TOKEN = os.environ.get("SODA_DALLAS_APPTOKEN")
DOMAIN = "www.dallasopendata.com"
DATASET_ID = "tbnj-w5hb"
DATASET_ID2 = "qqc2-eivj" #new verison 2.1 API no limits
```

```python
%%time
client = Socrata(DOMAIN, APP_TOKEN)
results = client.get(DATASET_ID2, limit=500000)
```

```python
len(results) #if greater than 438664 then we have all results
```
